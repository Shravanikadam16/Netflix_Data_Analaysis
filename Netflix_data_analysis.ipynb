# Exploratory Data Analysis (EDA)

## Movies vs TV Shows

```python
plt.figure(figsize=(6,4))

sns.countplot(x='type', data=df)

plt.title("Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Count")

plt.show()
```

## Insight

- Netflix has significantly more Movies than TV Shows.
- Movies account for the majority of the content available on the platform.
- This indicates that Netflix has historically focused more on movies than TV series.

```python
import seaborn as sns
```

## Rating Distribution

```python
plt.figure(figsize=(7,5))

sns.countplot(
    x='type',
    data=df,
    palette='Set2'
)

plt.title("Distribution of Netflix Content", fontsize=15)
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

## Insight

- TV-MA is the most common content rating.
- Netflix primarily offers content intended for mature audiences.

## Top 10  Countries

```python
top_country = df['country'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(x=top_country.values,
            y=top_country.index)

plt.title("Top 10 Countries on Netflix")

plt.show()
```

## Insight

- The United States has the highest number of Netflix titles.
- India and the United Kingdom are among the top content-producing countries.
- Netflix offers content from many countries, showing its global reach.

## Realese Year trend

```python
year = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,5))

plt.plot(year.index, year.values)

plt.title("Netflix Content Released by Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles")

plt.show()
```

## Insight

- The number of Netflix titles increased rapidly after 2015.
- This growth reflects Netflix's expansion in original and licensed content.

## Top 10 Genres

```python
genres = df['listed_in'].str.split(', ').explode()

top_genres = genres.value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(x=top_genres.values,
            y=top_genres.index)

plt.title("Top 10 Genres")

plt.show()
```

## Insight

- International Movies and Dramas are the most popular genres.
- Netflix provides a diverse range of content across different categories.

## Movie Duration

```python
movies = df[df['type'] == 'Movie'].copy()

movies['duration'] = movies['duration'].str.replace(' min','').astype(int)

plt.figure(figsize=(10,5))

sns.histplot(movies['duration'], bins=20)

plt.title("Movie Duration Distribution")

plt.show()
```

## Insight

- Most movies have a duration between 80 and 120 minutes.
- Feature-length movies dominate the Netflix movie catalog.

## Top Directors

```python
top_director = df['director'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(x=top_director.values,
            y=top_director.index)

plt.title("Top 10 Directors")

plt.show()
```

## Insight

- A small number of directors have multiple titles on Netflix.
- This suggests that some directors have a strong presence on the platform.
